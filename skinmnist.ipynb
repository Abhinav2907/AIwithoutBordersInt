```
from fastai import *
from fastai.vision import *
from fastai.callbacks.hooks import *

import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import auc,roc_curve

import os
```

```
class customlist(ImageList):
    def customlabelsgen(self,d, **kwargs)->'LabelList':
        """Custom Labels from path"""
        file_names=np.vectorize(lambda files: str(files).split('/')[-1][:-4])
        get_labels=lambda x: d.loc[x,'type']
        labels= get_labels(file_names(self.items))
        y = CategoryList(items=labels)
        res = self._label_list(x=self,y=y)
        return res#creating a databumch appropriate enough
```

```
def get_data(bs, size):
    train_ds = (customlist.from_folder('data/', extensions='.jpg')
                    .random_split_by_pct(0.15)
                    .customlabelsgen(d)
                    .transform(tfms=get_transforms(flip_vert=True),size=size)
                    .databunch(num_workers=2, bs=bs)
                    .normalize(imagenet_stats))
    return train_ds#creating databunch
```

```
data=get_data(16,224)
```

```
data.classes=list(np.unique(d.type))  
data.c= len(np.unique(d.type))  
```

```
data.show_batch(rows=3)
```

```

learner=cnn_learner(data,models.resnet50,metrics=[accuracy])
```

```
learner.lr_find()
learner.recorder.plot()#finding learning rate
```

```
learner.fit_one_cycle(7, 4e-3)#learnign rate talen to be somewhere between 1e-3 and 1e-2
```

```

learner1=cnn_learner(data,models.resnet50,metrics=[accuracy])# using transfer learning and taking resnet 50 architecture
```
